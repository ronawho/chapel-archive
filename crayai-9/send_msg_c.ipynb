```python
import time
import zmq
from  multiprocessing import Process
import sys
import datetime
import ctypes

lib_file = '/Users/eronagha/chapel/crayai-9/libsend_msg.so'
send_func = ctypes.CDLL(lib_file)

def send():
    print("SEND() PYTHON: calling c func")
    send_func.sendMsgC()

def mydate():
    date = datetime.datetime.utcnow()
    return f'{date.minute}:{date.second:02}'
    
def recv():
    context = zmq.Context()
    rep_socket = context.socket(zmq.REP)
    rep_socket.bind("tcp://*:5558")
    print(mydate(), " RECV(): in recv()")
    for num_msg in range(2):
        message = rep_socket.recv_string()
        print(mydate(), ' RECV(): recv msg: ', message, flush=True)
        rep_socket.send_string(message)
```

```python
Process(target=recv).start()
```
